```python
import pandas as pd
from datetime import datetime
```

```python
dataDir='../data/WATER_TEMP/'
df = pd.DataFrame()
for year in range(2015, 2020):  
    tmp_df = pd.read_csv(f'{dataDir}/glsea-temps{year}_1024.dat', skiprows=[0,1,2,3,4,5,6,8,9], delim_whitespace=True) 
    
    if len(df.values) == 0:
        df = tmp_df
    else:
        df=df.append(tmp_df)

#df.to_csv('GreatLakesWaterTemp_2015-2019', index=False) 
years = df['Year'].values
days = df['Day'].values
dates = []

for year, day in zip(years, days):
    dates.append(datetime.strptime(f'{year}_{day}', '%Y_%j'))

df.insert(0, 'Date', dates)
df=df.drop(['Year', 'Day'], axis=1)
df=df.rename(columns={'Sup.':'Superior Lake', 'Mich.':'Lake Michigan', 'Huron':'Lake Huron', 'Erie':'Lake Erie', 'Ont.':'Lake Ontario', 'St.Clr':'Lake Saint Clair'})
df
```
